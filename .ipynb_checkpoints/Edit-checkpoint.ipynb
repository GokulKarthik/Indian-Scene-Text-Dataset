```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_fp = "Gokul-Karthik-1.png"
image = np.array(Image.open(image_fp))
plt.imshow(image)
```

```python
white_image = np.ones(np.array(image).shape).astype('uint8')*255
plt.imshow(white_image)
```

```python
background_mask = np.array(image)[:,:,3]
print(background_mask.shape)
```

```python
transformed_image = image.copy()
transformed_image[np.where(background_mask == 0)] = white_image[np.where(background_mask == 0)]
```

```python
transformed_image = cv2.bitwise_or(np.array(white_image), np.array(image))
plt.imshow(transformed_image)
```

```python
cv2.imwrite("transformed.png", transformed_image)
```
